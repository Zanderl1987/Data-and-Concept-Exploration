# Note on the below try except statement.  
You need to have the entire code block within the try except statement because it runs through that first before going to any new code. So for any code outside of the try except statement, you'll get errors telling you that certain variables are being references prior to being defined.

```python
def GetSimilarCoData(sheetname):
    # Cannot use co_name as sheetname because they are different
    # co_name must be the CIQ short company name while the tabs contain the long company names
    
    #company_names_df = pd.read_excel("C:/Users/zanderl/Downloads/Analyst's Combined Watch List with CIQ IDs.xlsx",
    #                      encoding='utf-8',sheetname=2)
    
    try:
        df_company = pd.read_excel("C:/Users/zanderl/Documents/Data Science/Kensho Smart Connections Data/companies.xlsx",
                              encoding='utf-8',
                              sheetname=sheetname)

        co_name = str(df_company.iloc[sheetname,1])
        df_co = df_company[df_company['Company'] == str(co_name)]
        df_co.set_index(['Connection Type','Market Impact'])
        df_co_counts = df_co.pivot_table(index=['Market Impact','Connection Type'],aggfunc='count')
        df_co_counts.to_excel(co_name+' Own Company.xlsx')
        return [co_name, df_co_counts]
    except IndexError:
        pass
```

```python
for i in range(25,79,1):
    CoData_df = GetSimilarCoData(i)

try:
    CoData_df[0]
    CoData_df[1]
except TypeError:
    pass
#CoData_df
```
